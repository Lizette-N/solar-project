```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
figurstr = (15, 7)

t = np.linspace(0, 2 * np.pi, 1000)
f = np.cos(t)

plt.figure(figsize=figurstr)
plt.plot(t, f, color='b', linestyle='-')
plt.xlabel('t')
plt.ylabel('cos(t)')
plt.show()

# print(t,f)  # uncomment to see the values of t and f
```

```python
def InInterval(f, start, end):
    outf = f[f >= start] # Begrnser Intervallet til funktionsvrdier strre eller lig med intervalstarten
    outf = outf[outf <= end] # Begrnser Intervallet til funktionsvrdier mindre eller lig med intervalstarten
    return outf
    
ftest, ttest = InInterval(f,-0.5,0.5)
display(ftest, ttest)
```

```python
def find0Points(f):
    points = []
    for i in range(0,len(f)-1):
        if ((f[i] < 0 and f[i+1] > 0) or (f[i] > 0 and f[i+1] < 0)):
            points.append(f[i])
    return points

find0Points(f)
```
